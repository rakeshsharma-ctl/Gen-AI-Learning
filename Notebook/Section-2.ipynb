## Section 2: Embeddings — Semantic Space

### Why This Matters

After tokenization, each token is converted to an **embedding** — a dense vector of floating-point numbers (typically 1536 or 3072 dimensions) that encodes **meaning**.

- Vectors that are **close together** in this space = semantically similar concepts
- Vectors that are **far apart** = unrelated concepts
- This is what powers **semantic search**, **RAG pipelines**, and **document clustering**

We measure similarity with **cosine similarity**: ranges from -1 (opposite) to 1 (identical).

---

```python
# ─────────────────────────────────────────────────────────────
# HELPER: Cosine similarity between two vectors
#
# Cosine similarity = dot(A, B) / (||A|| * ||B||)
# Result: 1.0 = identical direction, 0.0 = orthogonal, -1.0 = opposite
# ─────────────────────────────────────────────────────────────

def cosine_similarity(vec_a: list, vec_b: list) -> float:
    a = np.array(vec_a)
    b = np.array(vec_b)
    return float(np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b)))


def get_embedding(text: str, model: str = "text-embedding-3-small") -> list:
    """
    Calls the OpenAI Embeddings API and returns the embedding vector.
    text-embedding-3-small: 1536 dimensions, fast, cheap
    text-embedding-3-large: 3072 dimensions, more accurate
    """
    response = client.embeddings.create(input=text, model=model)
    return response.data[0].embedding


print("Embedding helpers defined")
```

```python
# ─────────────────────────────────────────────────────────────
# DEMO: Three pairs of sentences
# We test three relationship types:
#   Pair A — Semantically similar (different words, same meaning)
#   Pair B — Syntactically similar, semantically different
#   Pair C — Completely unrelated
#
# Expected results:
#   Pair A: high similarity (> 0.8)
#   Pair B: moderate (0.4–0.7)
#   Pair C: low (< 0.3)
# ─────────────────────────────────────────────────────────────

pairs = [
    (
        "The model failed to retrieve the correct clause.",
        "The LLM ignored the relevant section of the contract.",
        "Pair A — Semantically similar"
    ),
    (
        "The bank processed the financial transaction quickly.",
        "The river bank flooded during the heavy rainfall.",
        "Pair B — Same word 'bank', different meaning"
    ),
    (
        "Transformer attention weights are computed per token.",
        "The quarterly revenue exceeded analyst expectations.",
        "Pair C — Unrelated topics"
    ),
]

print("Generating embeddings and computing similarities...\n")
for s1, s2, label in pairs:
    emb1 = get_embedding(s1)
    emb2 = get_embedding(s2)
    sim = cosine_similarity(emb1, emb2)
    print(f"{label}")
    print(f"  S1: {s1}")
    print(f"  S2: {s2}")
    print(f"  Cosine Similarity: {sim:.4f}")
    print()
```
